```python
%matplotlib widget
import fastf1
import fastf1.core
import fastf1.events
from fastf1.plotting import get_compound_color, get_driver_color, list_compounds
import matplotlib.pyplot as plt
```

```python
session = fastf1.get_session(2025, 'China', 'Q')
session.load()

session
```

```python
ver_laps = session.laps.pick_drivers('VER').pick_fastest()
law_laps = session.laps.pick_drivers('LAW').pick_fastest()
tsu_laps = session.laps.pick_drivers('TSU').pick_fastest()

ver_telem = ver_laps.telemetry[ver_laps.telemetry['Source'] == 'pos']
law_telem = law_laps.telemetry[law_laps.telemetry['Source'] == 'pos']
tsu_telem = tsu_laps.telemetry[tsu_laps.telemetry['Source'] == 'pos']
```

```python
fig, ax = plt.subplots()

ax.plot(ver_telem['Time'], ver_telem['Speed'], label = 'VER')
ax.plot(law_telem['Time'], law_telem['Speed'], label = 'LAW')
# ax.plot(tsu_telem['Time'], tsu_telem['Speed'], label = 'TSU')

ax.set_xlabel('Time')
ax.set_ylabel('Speed')
ax.set_facecolor('#272A37')
ax.legend()

plt.ion()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(ver_telem['Time'], ver_telem['Brake'], label = 'VER')
ax.plot(law_telem['Time'], law_telem['Brake'], label = 'LAW')
# ax.plot(tsu_telem['Time'], tsu_telem['Brake'], label = 'TSU')

ax.set_xlabel('Time')
ax.set_ylabel('Speed')
ax.set_facecolor('#272A37')
ax.legend()

plt.ion()
plt.show()
```
